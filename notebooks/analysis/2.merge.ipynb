```python
import pandas as pd
import glob
import os
from pathlib import Path

PROCESSED_DIR = Path.cwd() / "data" / "processed" / "left_right"
DATA_ATTACK_DIR = PROCESSED_DIR / "data_attack"

def merge_all_attack_data(output_filename="all_hitters_historical.csv"):
    # 1. Get a list of all CSV files in the data_attack folder
    all_files = sorted(DATA_ATTACK_DIR.glob("*.csv"))
    
    if not all_files:
        print("No CSV files found in data_attack!")
        return

    # 2. Read each file and put them into a list
    df_list = []
    for filename in all_files:
        # We read the file and add it to our list
        df = pd.read_csv(filename)
        df_list.append(df)
        print(f"Loaded {filename} with {len(df)} rows.")

    # 3. Concatenate all DataFrames into one
    master_df = pd.concat(df_list, axis=0, ignore_index=True)

    # 4. Save the massive file
    output_path = PROCESSED_DIR / output_filename
    master_df.to_csv(output_path, index=False)
    print(f"\nSuccessfully merged {len(all_files)} files!")
    print(f"Total rows in master file: {len(master_df)}")
    print(f"Saved as: {output_path}")

# Run the merge
merge_all_attack_data()
```
